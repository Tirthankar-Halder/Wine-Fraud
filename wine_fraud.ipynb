### Collecting the requred features for training

```python
X=pd.DataFrame(wine,columns=wine['quality'])
Y=pd.DataFrame(wine,columns=wine['cost'])
```

```python
#Split your data into training and testing sets using SciKit Learn's train_test_split function from model_selection
Xtrain,Xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3,random_state=6)
```
