# Sweetviz Automated EDA : Wine Quality

```python
#importing required module for this project
import pandas as pd
import sweetviz as sv
```

```python
#loading Dataset 
wine= pd.read_csv('wine.csv')
```

```python
#checking the data
wine.head()
```

```python
column_names=['Type','Fixed Acidity','Volatile Acidity','Citric Acid','Residual Sugar','Chlorides','Free Sulfur Dioxide','Total Sulfur Dioxide','Density','pH','Sulphates','Alcohol','Quality','Cost']
wine.columns=column_names
```

```python
wine.head(10)
```

```python
#checking the shape of the dataset
wine.shape
```

## Analyzing the Dataset 

```python
report= sv.analyze(wine)
```

```python
report.show_html('Analyze_report.html')
```
